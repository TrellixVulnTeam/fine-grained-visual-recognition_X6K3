```python
train_df = pd.read_csv(DATA_DIR + 'DF20-train_metadata_PROD.csv')
```

## Image Examples

```python
dataset = data.ImageDataset(
    train_df, path=DATA_DIR + TRAIN_SET_DIR,
    img_path_col='image_path', label_col='scientificName')
```

```python
idx = [194201, 103728, 33069, 40662, 96810, 161171, 250854, 78449]  # 231338
dataset.show_items(idx, ncols=4)
```

```python
import re

for i, _idx in enumerate(idx):
    img, label = dataset.get_item(_idx)
    img = img.resize((224, 224))
    filename = label.replace(' ', '_').replace('.', '').lower()
    filename = re.sub('_\(.*', '', filename)
    img.save(f'figures/fungi_examples/{filename}.png')
```
