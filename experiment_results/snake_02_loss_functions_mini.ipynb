```python
from src.utils import visualization as viz
from src.dev import results

DATA_DIR = 'data/snake_clef2021_dataset/training_history/loss_functions_mini/'


# load specification and progress data
specs_df = results.load_specs_files(DATA_DIR)
# specs_df['note'] = specs_df['note'].fillna('')
df = results.load_progress_files(specs_df, path=DATA_DIR)
specs_df
```

## EfficientNet-B0

```python
params = dict(model='efficientnet_b0', total_batch_size=64)
group_dict = {
    'CE': results.filter_items(df, outlen=30, loss='ce', weight='none', **params),
    'CE - Inverse Class Frequency': results.filter_items(df, outlen=30, loss='ce', weight='inverse_class_frequency', **params),
    'CE - Linear Class': results.filter_items(df, outlen=30, loss='ce', weight='linear_class', **params),
    'CE - Modified Linear Class': results.filter_items(df, outlen=30, loss='ce', weight='modified_linear_class', **params),
    'CE - Class Balanced (beta=0.9)': results.filter_items(df, outlen=30, loss='ce', weight='class_balanced', beta=0.9, **params),
    'CE - Class Balanced (beta=0.99)': results.filter_items(df, outlen=30, loss='ce', weight='class_balanced', beta=0.99, **params),
    'CE - Class Balanced (beta=0.999)': results.filter_items(df, outlen=30, loss='ce', weight='class_balanced', beta=0.999, **params),
    'CE - Class Balanced (beta=0.9999)': results.filter_items(df, outlen=30, loss='ce', weight='class_balanced', beta=0.9999, **params),
    'FL (gamma=0.5)': results.filter_items(df, outlen=None, loss='focal', weight='none', gamma=0.5, **params),
    'FL (gamma=1.0)': results.filter_items(df, outlen=30, loss='focal', weight='none', gamma=1.0, **params),
    'FL (gamma=2.0)': results.filter_items(df, outlen=30, loss='focal', weight='none', gamma=2.0, **params),
    'FL (gamma=5.0)': results.filter_items(df, outlen=30, loss='focal', weight='none', gamma=5.0, **params)}

result_df = results.get_metrics_and_time_df(group_dict, primary_metric='f1_score')
viz.heatmap(result_df)
result_df.round(3) * ([100] * 3 + [1, 1])
```

```python
print(f'FL (gamma=0.5) trained only for: {len(group_dict["FL (gamma=0.5)"])}')
```

```python
_result_df = result_df[['accuracy', 'top_3', 'f1_score']]
(_result_df - _result_df.loc['CE']).round(3) * 100
```
