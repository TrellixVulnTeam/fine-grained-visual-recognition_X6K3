```python
import numpy as np
import pandas as pd

from src.utils import visualization as viz
from src.dev import results

DATA_DIR = 'data/danish_fungi_dataset/training_history/image_cropping_mini/'


# load specification and progress data
specs_df = results.load_specs_files(DATA_DIR)
specs_df['note'] = specs_df['note'].fillna('')
df = results.load_progress_files(specs_df, path=DATA_DIR)
specs_df['history_file_full'] = specs_df['model_name'] + '_full.csv'
df_full = results.load_progress_files(specs_df, path=DATA_DIR, history_file_col='history_file_full')
specs_df
```

```python
def _filter_items(df_crop, df_full, outlen, **kwargs):
    group_crop = results.filter_items(df_crop, outlen=outlen, **kwargs)
    group_full = results.filter_items(df_full, outlen=1, **kwargs)
    return group_crop, group_full


def expand_group_tuple(group_crop, group_full, inverse_order=False):
    if inverse_order:
        pref1 = '_full'
        pref2 = '_crop'
        group_crop, group_full = group_full, group_crop
    else:
        pref1 = '_crop'
        pref2 = '_full'
    o = results.get_metrics_and_time(group_crop)
    metrics = ['accuracy', 'top_3', 'f1_score']
    for met in metrics:
        o[f'{met}{pref1}'] = o.pop(met)
    o.update({f'{k}{pref2}': v for k, v in group_full[metrics].iloc[0].items()})
    return o


params = dict(loss='ce', weight='class_balanced', beta=0.9999, opt='sgd', model='efficientnet_b0', total_batch_size=64)

g1_full = results.filter_items(df, outlen=30, note='', **params)
g1_tuple = pd.DataFrame([{'accuracy': 0.5903846153846154, 'top_3': 0.7865384615384615, 'f1_score': 0.487755128522515}]), g1_full

g2_tuple = _filter_items(df, df_full, outlen=30, note='trained on crop', **params)
g3_tuple = _filter_items(df, df_full, outlen=15, note='trained on full and crop', **params)
g4_tuple = _filter_items(df, df_full, outlen=30, note='trained on full and crop (random)', **params)
g5 = results.filter_items(df, outlen=30, note='trained on full and crop (6 channels)', **params)
g5_tuple = g5, pd.DataFrame([{'accuracy': None, 'top_3': None, 'f1_score': None}])

g6_tuple = _filter_items(df, df_full, outlen=30, note='trained on full and crop (mixup)', alpha=0.2, mixup_p=0.25, **params)
g7_tuple = _filter_items(df, df_full, outlen=30, note='trained on full and crop (mixup)', alpha=0.5, mixup_p=0.25, **params)
g8_tuple = _filter_items(df, df_full, outlen=30, note='trained on full and crop (mixup)', alpha=0.8, mixup_p=0.25, **params)

g9_tuple = _filter_items(df, df_full, outlen=30, note='trained on full and crop (mixup)', alpha=0.2, mixup_p=0.5, **params)
g10_tuple = _filter_items(df, df_full, outlen=30, note='trained on full and crop (mixup)', alpha=0.5, mixup_p=0.5, **params)
g11_tuple = _filter_items(df, df_full, outlen=30, note='trained on full and crop (mixup)', alpha=0.8, mixup_p=0.5, **params)

# g12_tuple = _filter_items(df, df_full, outlen=30, note='trained on full and crop (mixup)', alpha=0.2, mixup_p=0.75, **params)
# g13_tuple = _filter_items(df, df_full, outlen=30, note='trained on full and crop (mixup)', alpha=0.5, mixup_p=0.75, **params)
# g14_tuple = _filter_items(df, df_full, outlen=30, note='trained on full and crop (mixup)', alpha=0.8, mixup_p=0.75, **params)


group_dict = {
    'Trained on crop': g2_tuple,
    'Trained on full and crop': g3_tuple,
    'Trained on full and crop (random)': g4_tuple,
    'Trained on full and crop (6 channels)': g5_tuple,

    'Trained on full and crop (mixup: alpha=0.2, p=0.25)': g6_tuple,
    'Trained on full and crop (mixup: alpha=0.5, p=0.25)': g7_tuple,
    'Trained on full and crop (mixup: alpha=0.8, p=0.25)': g8_tuple,

    'Trained on full and crop (mixup: alpha=0.2, p=0.5)': g9_tuple,
    'Trained on full and crop (mixup: alpha=0.5, p=0.5)': g10_tuple,
    'Trained on full and crop (mixup: alpha=0.8, p=0.5)': g11_tuple,

    # 'Trained on full and crop (mixup: alpha=0.2, p=0.75)': g12_tuple,
    # 'Trained on full and crop (mixup: alpha=0.5, p=0.75)': g13_tuple,
    # 'Trained on full and crop (mixup: alpha=0.8, p=0.75)': g14_tuple,
}
_results = {'Trained on full': expand_group_tuple(*g1_tuple, inverse_order=True)}
for k, (group_crop, group_full) in group_dict.items():
    _results[k] = expand_group_tuple(group_crop, group_full)
result_df = pd.DataFrame.from_dict(_results, orient='index')
result_df = result_df[['accuracy_full', 'accuracy_crop', 'top_3_full', 'top_3_crop', 'f1_score_full', 'f1_score_crop',
                       'mean_epoch_time', 'total_time']]
viz.heatmap(result_df)

result_df.round(3) * ([100] * 6 + [1, 1])
```
